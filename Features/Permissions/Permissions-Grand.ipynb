```
import os
from pyaxmlparser import APK
```

```
def process(arr1):
  res = []
  for i in arr1:
    temp = i.split('.')
    if 'permission' in temp:
      ind = temp.index('permission')
      res.append(".".join(temp[ind:]))
  return res
```

```
#path-check

#Malware Files
malware_count = 0
list1 = []
for entry in os.scandir("/content/drive/MyDrive/Project BE 2020-2021/Datasets/ZenodoMerged/"):
    if malware_count == 100:
        break
    try:    
        if entry.path.endswith(".apk") and entry.is_file():
            a = APK(entry.path)
            list1.append(process(a.get_permissions()))
            malware_count += 1
            print(malware_count)         
    except:
        print(entry.path)
print(list1)
```

```
#path-check

#Benign Files
benign_count = 0
for entry in os.scandir("/content/drive/MyDrive/Project BE 2020-2021/Datasets/Benign_2015_2016_2017/"):
    if benign_count == 100:
        break
    try:    
        if entry.path.endswith(".apk") and entry.is_file():
            a = APK(entry.path)
            list1.append(process(a.get_permissions()))
            benign_count += 1
            print(benign_count)         
    except:
        print(entry.path)
print(list1)
```

```
allPermissions = set()
for i in list1:
    allPermissions = allPermissions.union(set(i))
print(allPermissions)
```

```
print(len(allPermissions))

allPermissions = list(allPermissions)
```

```
rowList = []
for i in range(len(list1)):
    x = []
    for j in range(len(allPermissions)):
        x.append(0)
    rowList.append(x)
```

```
for i in range(len(list1)):
    for j in list1[i]:
        x = allPermissions.index(j)
        if x > -1:
            rowList[i][x] = 1
```

```
allPermissions.append('class')
for i in range(0, malware_count):
    rowList[i].append(1)
while(True):
    i += 1
    if i == len(rowList):
        break
    rowList[i].append(0)
```

#MODEL PREPARATION

```
from __future__ import print_function
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import tensorflow as tf
import keras
np.random.seed(1337)  # for reproducibility
from keras.utils import np_utils
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Embedding
from keras.layers import LSTM, SimpleRNN, GRU
from keras.datasets import imdb
from keras.utils.np_utils import to_categorical
from sklearn.metrics import (precision_score, recall_score,f1_score, accuracy_score,mean_squared_error,mean_absolute_error)
from sklearn import metrics
from sklearn.preprocessing import Normalizer
import h5py
from keras import callbacks
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau, CSVLogger
```

```
#path-check

traindata = pd.read_csv('/content/drive/MyDrive/Project BE 2020-2021/Quarter Final/Permissions/Permission Dataset.csv')
testdata = pd.read_csv('/content/drive/MyDrive/Project BE 2020-2021/Quarter Final/Permissions/Permission Dataset.csv')
```

```
traindata.shape
```

```
X = traindata.iloc[1:,0:(traindata.shape[1]-1)]
Y = traindata.iloc[1:,(traindata.shape[1]-1)]
C = testdata.iloc[1:,(traindata.shape[1]-1)]
T = testdata.iloc[1:,0:(traindata.shape[1]-1)]
```

```
scaler = Normalizer().fit(X)
trainX = scaler.transform(X)
np.set_printoptions(precision=3)
```

```
scaler = Normalizer().fit(T)
testT = scaler.transform(T)
np.set_printoptions(precision=3)
```

```
y_train = np.array(Y)
y_test = np.array(C)
```

```
X_train = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
X_test = np.reshape(testT, (testT.shape[0], 1, testT.shape[1]))
```

```
batch_size = 32
```

```
model = Sequential()
model.add(LSTM(64,input_dim=(traindata.shape[1]-1), return_sequences=True))
model.add(Dropout(0.1))
model.add(LSTM(64, return_sequences=True))
model.add(Dropout(0.1))
model.add(LSTM(64, return_sequences=True))
model.add(Dropout(0.1))
model.add(LSTM(64, return_sequences=True))
model.add(Dropout(0.1))
model.add(LSTM(64, return_sequences=True)) 
model.add(Dropout(0.1))
model.add(LSTM(64, return_sequences=False))
model.add(Dropout(0.1))
model.add(Dense(1))
model.add(Activation('sigmoid'))
model.summary()
```

```
#epoch-check
#path-check

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train, y_train, batch_size=batch_size, epochs=10, validation_data=(X_test, y_test))
model.save("PermissionModel.hdf5")
```
